## üìò –§–æ—Ä–º—É–ª–∏—Ä–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏

–ü—Ä–µ–¥—Å—Ç–∞–≤—å—Ç–µ, —á—Ç–æ –≤—ã —Ä–∞–±–æ—Ç–∞–µ—Ç–µ –∞–Ω–∞–ª–∏—Ç–∏–∫–æ–º –≤ HR-–æ—Ç–¥–µ–ª–µ –∫–æ–º–ø–∞–Ω–∏–∏. –í–∞–º –Ω—É–∂–Ω–æ —Å–º–æ–¥–µ–ª–∏—Ä–æ–≤–∞—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–∞—Ä–ø–ª–∞—Ç —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –≤ —Ç—Ä—ë—Ö –æ—Ç–¥–µ–ª–∞—Ö (IT, Sales, HR) –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—é –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è.

### –ò—Å–ø–æ–ª—å–∑—É—è numpy, —Å–≥–µ–Ω–µ—Ä–∏—Ä—É–π—Ç–µ –≤—ã–±–æ—Ä–∫–∏ –∑–∞—Ä–ø–ª–∞—Ç:

**IT-–æ—Ç–¥–µ–ª: —Å—Ä–µ–¥–Ω—è—è –∑–∞—Ä–ø–ª–∞—Ç–∞ ‚âà 120 000 —Ä—É–±., —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ 20 000 —Ä—É–±.**

**Sales-–æ—Ç–¥–µ–ª: —Å—Ä–µ–¥–Ω—è—è –∑–∞—Ä–ø–ª–∞—Ç–∞ ‚âà 80 000 —Ä—É–±., —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ 15 000 —Ä—É–±.**

**HR-–æ—Ç–¥–µ–ª: —Å—Ä–µ–¥–Ω—è—è –∑–∞—Ä–ø–ª–∞—Ç–∞ ‚âà 60 000 —Ä—É–±., —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ 10 000 —Ä—É–±.**

**–ö–∞–∂–¥–∞—è –≤—ã–±–æ—Ä–∫–∞ ‚Äî 200 —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤.**

### –ü–æ—Å—Ç—Ä–æ–π—Ç–µ —Å –ø–æ–º–æ—â—å—é matplotlib:

**–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∑–∞—Ä–ø–ª–∞—Ç –¥–ª—è —Ç—Ä—ë—Ö –æ—Ç–¥–µ–ª–æ–≤ –Ω–∞ –æ–¥–Ω–æ–º –≥—Ä–∞—Ñ–∏–∫–µ.**

**–ö–æ—Ä–æ–±—á–∞—Ç—É—é –¥–∏–∞–≥—Ä–∞–º–º—É (boxplot) –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è.**

### –í—ã—á–∏—Å–ª–∏—Ç–µ –∏ –≤—ã–≤–µ–¥–∏—Ç–µ:

**—Å—Ä–µ–¥–Ω–µ–µ, –º–µ–¥–∏–∞–Ω—É –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –æ—Ç–¥–µ–ª–∞.**

**–æ—Ç–¥–µ–ª —Å –Ω–∞–∏–±–æ–ª—å—à–µ–π –≤–∞—Ä–∏–∞—Ç–∏–≤–Ω–æ—Å—Ç—å—é –∑–∞—Ä–ø–ª–∞—Ç.**

### üîπ Python (–±–∞–∑–æ–≤—ã–µ –≤–µ—â–∏)

zip() ‚Äî –æ–±—ä–µ–¥–∏–Ω—è–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ —Å–ø–∏—Å–∫–æ–≤/–º–∞—Å—Å–∏–≤–æ–≤ –≤ –ø–∞—Ä—ã –¥–ª—è –ø–µ—Ä–µ–±–æ—Ä–∞.

–¶–∏–∫–ª—ã for, —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ —Å—Ç—Ä–æ–∫ f"...".

–†–∞–±–æ—Ç–∞ —Å–æ —Å–ø–∏—Å–∫–∞–º–∏ –∏ –º–∞—Å—Å–∏–≤–∞–º–∏.

### üîπ NumPy (–∫–ª—é—á–µ–≤—ã–µ –º–µ—Ç–æ–¥—ã)

np.random.normal(mean, std, size) ‚Äî –≥–µ–Ω–µ—Ä–∞—Ü–∏—è —Å–ª—É—á–∞–π–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö (–Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ).

np.mean(arr) ‚Äî —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ.

np.median(arr) ‚Äî –º–µ–¥–∏–∞–Ω–∞.

np.std(arr) ‚Äî —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ (—Ä–∞–∑–±—Ä–æ—Å –¥–∞–Ω–Ω—ã—Ö).

np.arange(start, stop, step) –∏ np.linspace(start, stop, num) ‚Äî —Å–æ–∑–¥–∞–Ω–∏–µ –º–∞—Å—Å–∏–≤–æ–≤.

–ò–Ω–¥–µ–∫—Å–∞—Ü–∏—è –∏ —Å—Ä–µ–∑—ã –º–∞—Å—Å–∏–≤–æ–≤ (–Ω–∞–ø—Ä–∏–º–µ—Ä, arr[0:10]).

### üîπ Matplotlib (–º–∏–Ω–∏–º—É–º)

plt.hist() ‚Äî –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞.

plt.boxplot() ‚Äî boxplot.

plt.plot() ‚Äî –ª–∏–Ω–µ–π–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫.

plt.show() ‚Äî –ø–æ–∫–∞–∑–∞—Ç—å –≥—Ä–∞—Ñ–∏–∫.

```python
# –ü–æ–¥–∫–ª—é—á–∞–µ–º –±–∏–±–ª–∏–æ—Ç–µ–∫–∏:
# numpy ‚Äî –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –º–∞—Å—Å–∏–≤–∞–º–∏ –∏ –≥–µ–Ω–µ—Ä–∞—Ü–∏–µ–π —Å–ª—É—á–∞–π–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö,
# matplotlib.pyplot ‚Äî –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤.
import numpy as np
import matplotlib.pyplot as plt

# –§–∏–∫—Å–∏—Ä—É–µ–º "–∑–µ—Ä–Ω–æ" –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞ —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª,
# —á—Ç–æ–±—ã –∫–∞–∂–¥—ã–π –∑–∞–ø—É—Å–∫ –∫–æ–¥–∞ –¥–∞–≤–∞–ª –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã (—É–¥–æ–±–Ω–æ –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –∏ –æ—Ç–ª–∞–¥–∫–∏).
np.random.seed(42)

# –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –∑–∞—Ä–ø–ª–∞—Ç—ã –≤ IT-–æ—Ç–¥–µ–ª–µ:
# —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ (–∫–æ–ª–æ–∫–æ–ª), —Å—Ä–µ–¥–Ω–µ–µ = 120 000, —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ = 20 000,
# –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ = 200.
it_salaries = np.random.normal(120_000, 20_000, 200)

# –ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –∑–∞—Ä–ø–ª–∞—Ç—ã –≤ Sales-–æ—Ç–¥–µ–ª–µ: —Å—Ä–µ–¥–Ω–µ–µ 80 000, —Ä–∞–∑–±—Ä–æ—Å 15 000.
sales_salaries = np.random.normal(80_000, 15_000, 200)

# –ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ –¥–ª—è HR-–æ—Ç–¥–µ–ª–∞: —Å—Ä–µ–¥–Ω–µ–µ 60 000, —Ä–∞–∑–±—Ä–æ—Å 10 000.
hr_salaries = np.random.normal(60_000, 10_000, 200)

# ---------- –í–ò–ó–£–ê–õ–ò–ó–ê–¶–ò–Ø ‚Ññ1: –ì–ò–°–¢–û–ì–†–ê–ú–ú–´ ----------

# –°–æ–∑–¥–∞—ë–º –æ–∫–Ω–æ –≥—Ä–∞—Ñ–∏–∫–∞ —Å —Ä–∞–∑–º–µ—Ä–æ–º 10 –Ω–∞ 6 –¥—é–π–º–æ–≤.
plt.figure(figsize=(10, 6))

# –°—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –∑–∞—Ä–ø–ª–∞—Ç –≤ IT, —Ä–∞–∑–±–∏–≤–∞–µ–º –¥–∏–∞–ø–∞–∑–æ–Ω –Ω–∞ 20 –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤ (bins),
# –¥–µ–ª–∞–µ–º –ø—Ä–æ–∑—Ä–∞—á–Ω–æ—Å—Ç—å 0.6, –¥–æ–±–∞–≤–ª—è–µ–º –ø–æ–¥–ø–∏—Å—å "IT".
plt.hist(it_salaries, bins=20, alpha=0.6, label="IT")

# –¢–æ –∂–µ –¥–ª—è Sales.
plt.hist(sales_salaries, bins=20, alpha=0.6, label="Sales")

# –¢–æ –∂–µ –¥–ª—è HR.
plt.hist(hr_salaries, bins=20, alpha=0.6, label="HR")

# –î–æ–±–∞–≤–ª—è–µ–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –≥—Ä–∞—Ñ–∏–∫–∞.
plt.title("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–∞—Ä–ø–ª–∞—Ç —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤")

# –ü–æ–¥–ø–∏—Å—ã–≤–∞–µ–º –æ—Å—å X ‚Äî —ç—Ç–æ –∑–∞—Ä–ø–ª–∞—Ç–∞ –≤ —Ä—É–±–ª—è—Ö.
plt.xlabel("–ó–∞—Ä–ø–ª–∞—Ç–∞, —Ä—É–±.")

# –ü–æ–¥–ø–∏—Å—ã–≤–∞–µ–º –æ—Å—å Y ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –≤ –∫–∞–∂–¥–æ–º –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ –∑–∞—Ä–ø–ª–∞—Ç.
plt.ylabel("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤")

# –î–æ–±–∞–≤–ª—è–µ–º –ª–µ–≥–µ–Ω–¥—É (—á—Ç–æ–±—ã —Ä–∞–∑–ª–∏—á–∞—Ç—å –æ—Ç–¥–µ–ª—ã –ø–æ —Ü–≤–µ—Ç–∞–º).
plt.legend()

# –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –ø–æ—Å—Ç—Ä–æ–µ–Ω–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫.
plt.show()

# ---------- –í–ò–ó–£–ê–õ–ò–ó–ê–¶–ò–Ø ‚Ññ2: BOXPLOT ----------

# –°–æ–∑–¥–∞—ë–º –æ–∫–Ω–æ –≥—Ä–∞—Ñ–∏–∫–∞ —Ä–∞–∑–º–µ—Ä–æ–º 8 –Ω–∞ 5 –¥—é–π–º–æ–≤.
plt.figure(figsize=(8, 5))

# –°—Ç—Ä–æ–∏–º boxplot (–∫–æ—Ä–æ–±—á–∞—Ç—É—é –¥–∏–∞–≥—Ä–∞–º–º—É) –¥–ª—è —Ç—Ä—ë—Ö –æ—Ç–¥–µ–ª–æ–≤.
# –≠—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç —Å—Ä–∞–≤–Ω–∏—Ç—å –º–µ–¥–∏–∞–Ω—É, –∫–≤–∞—Ä—Ç–∏–ª–∏ –∏ –≤—ã–±—Ä–æ—Å—ã.
plt.boxplot([it_salaries, sales_salaries, hr_salaries], labels=["IT", "Sales", "HR"])

# –î–æ–±–∞–≤–ª—è–µ–º –∑–∞–≥–æ–ª–æ–≤–æ–∫.
plt.title("–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –∑–∞—Ä–ø–ª–∞—Ç –ø–æ –æ—Ç–¥–µ–ª–∞–º (boxplot)")

# –ü–æ–¥–ø–∏—Å—ã–≤–∞–µ–º –æ—Å—å Y.
plt.ylabel("–ó–∞—Ä–ø–ª–∞—Ç–∞, —Ä—É–±.")

# –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –≥—Ä–∞—Ñ–∏–∫.
plt.show()

# ---------- –†–ê–°–ß–Å–¢ –°–¢–ê–¢–ò–°–¢–ò–ö ----------

# –ü–µ—Ä–µ–±–∏—Ä–∞–µ–º –ø–∞—Ä—ã "–Ω–∞–∑–≤–∞–Ω–∏–µ –æ—Ç–¥–µ–ª–∞" –∏ "–º–∞—Å—Å–∏–≤ –¥–∞–Ω–Ω—ã—Ö"
for dept, data in zip(["IT", "Sales", "HR"], [it_salaries, sales_salaries, hr_salaries]):
    # –í—ã–≤–æ–¥–∏–º —Å—Ä–µ–¥–Ω–µ–µ, –º–µ–¥–∏–∞–Ω—É –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ (–æ–∫—Ä—É–≥–ª—è–µ–º –¥–æ —Ü–µ–ª–æ–≥–æ)
    print(f"{dept}: –°—Ä–µ–¥–Ω–µ–µ={np.mean(data):.0f}, "
          f"–ú–µ–¥–∏–∞–Ω–∞={np.median(data):.0f}, "
          f"Std={np.std(data):.0f}")
```

–ø—Ä–æ—Ü–µ–Ω—Ç —Å—Ç—É–¥–µ–Ω—Ç–æ–≤, –ø—Ä–æ–≤–æ–¥—è—â–∏—Ö –≤ —Å–æ—Ü—Å–µ—Ç—è—Ö –±–æ–ª–µ–µ 4 —á–∞—Å–æ–≤ –≤ –¥–µ–Ω—å.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö
it_salaries = np.random.normal(120_000, 20_000, 200)
sales_salaries = np.random.normal(80_000, 15_000, 200)
hr_salaries = np.random.normal(60_000, 10_000, 200)
```

```python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º
plt.figure(figsize=(10, 6))
plt.hist(it_salaries, bins=20, alpha=0.6, label='IT')
plt.hist(sales_salaries, bins=20, alpha=0.6, label='Sales')
plt.hist(hr_salaries, bins=20, alpha=0.6, label='HR')
plt.legend()
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–∞—Ä–ø–ª–∞—Ç —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –ø–æ –æ—Ç–¥–µ–ª–∞–º')
plt.xlabel('–ó–∞—Ä–ø–ª–∞—Ç–∞, —Ä—É–±.')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤')
plt.show()
```

```python
# –ö–æ—Ä–æ–±—á–∞—Ç–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞
plt.figure(figsize=(8, 6))
plt.boxplot([it_salaries, sales_salaries, hr_salaries], tick_labels=['IT', 'Sales', 'HR'])
plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∑–∞—Ä–ø–ª–∞—Ç –≤ –æ—Ç–¥–µ–ª–∞—Ö')
plt.ylabel('–ó–∞—Ä–ø–ª–∞—Ç–∞, —Ä—É–±.')
plt.show()
```

```python
# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫
departments = {'IT': it_salaries, 'Sales': sales_salaries, 'HR': hr_salaries}
for dept, salaries in departments.items():
    mean = np.mean(salaries)
    median = np.median(salaries)
    std = np.std(salaries)
    print(f"{dept}: –°—Ä–µ–¥–Ω–µ–µ={mean:.0f}, –ú–µ–¥–∏–∞–Ω–∞={median:.0f}, Std={std:.0f}")

# –ù–∞–π—Ç–∏ –æ—Ç–¥–µ–ª —Å –Ω–∞–∏–±–æ–ª—å—à–µ–π –≤–∞—Ä–∏–∞—Ç–∏–≤–Ω–æ—Å—Ç—å—é (—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ–º)
max_variability_dept = max(departments.items(), key=lambda x: np.std(x[1]))[0]
print(f"\n–û—Ç–¥–µ–ª —Å –Ω–∞–∏–±–æ–ª—å—à–µ–π –≤–∞—Ä–∏–∞—Ç–∏–≤–Ω–æ—Å—Ç—å—é –∑–∞—Ä–ø–ª–∞—Ç: {max_variability_dept}")
```
